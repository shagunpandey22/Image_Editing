```python
def rotate_using_pil(img):
    img2=Image.rotate(img)
    pt.imshow(img2)
    pt.show()
```

```python

def rotates(img):
    img3=img.copy()
    '''pt.imshow(img3) 
    pt.show()'''
    #imgg=img.sum(2)/(255*3)  #image converted to grayscale as it is helpful for processing of segmentation
    f = pt.figure(figsize=(10, 10))
    f.add_subplot(1, 2, 1)
    pt.imshow(img)
    pt.title('The Original Image')
    d=90
    for _ in range(d//2):
        img3=img3.transpose(1,0,2)
        for j in range(0, img3.shape[1] // 2):
            c = img3[:, j, :].copy()
            img3[:, j, :] = img3[: , img3.shape[1]-j-1, :]
            img3[: , img3.shape[1]-j-1, :] = c
    f.add_subplot(1, 2, 2)
    pt.imshow(img3)
    pt.title("rotated")

    
    
```

```python
rotates(img)
```

```python
### Negative Image
To convert the image into negative image get the rgb values of the image and 
peroform 255-R,255-G,255-B
and save the rgb values in the RGB
```

```python
def negative(img):
    fig = pt.figure(figsize=(10, 10))
    img = 255 - img
    fig.add_subplot(1, 2, 2)
    pt.imshow(img)
    pt.title('Negative of RGB image')
```

```python
negative(img)
```

```python
def negative_grey(img4):
    fig = pt.figure(figsize=(10, 10))
    img_grey = 255*3 - img4        
    fig.add_subplot(1, 2, 1)
    pt.imshow(img_grey)
    pt.title('Negative of Grey image')
    
```

```python
negative_grey(img4)
```

```python
def rgb(img):
    img1=np.array(img) 
    img_R=img1.copy()
    img_G=img1.copy()
    img_B=img1.copy()
    img_R[:, :, (1, 2)] = 0
    img_G[:, :, (0, 2)] = 0
    img_B[:, :, (0, 1)] = 0
    img_rgb = np.concatenate((img_R,img_G,img_B), axis=1)
    pt.figure(figsize=(15, 15))
    pt.imshow(img_rgb)
```

```python
rgb(img)
```

```python
def trim(img,i):
    img1=img.copy()
    fig = pt.figure(figsize=(10, 10))
    fig.add_subplot(1, 2, 1)
    pt.imshow(img1)
    pt.title('Original')
    img_t= img[i:-i, i:-i, :]   #Slicing
    fig.add_subplot(1, 2, 2)
    pt.imshow(img_t)
    pt.title('Trimmed')
#Triming of image is done with the help of Slicing
```

```python
i=int(input("Enter The Size To Trim"))  #Taking the input from user
```

```python
trim(img,i)  #passing through the trim function the image and the size
```

```python
def modify(img):
    img1=img.copy()
    img2 = (img // 64) * 64
    img3 = (img // 128) * 128
    imgg = np.concatenate((img1, img2, img3), axis=1)
    pt.figure(figsize=(15, 15))
    pt.imshow(imgg)
# Making Pixel values discrete by first division by // which gives int and then multiply by the same factor
```

```python
modify(img)
```

```python
def flip(img):
    img1=img.copy()
    fig = pt.figure(figsize=(10, 10))
    fig.add_subplot(1, 2, 1)
    pt.imshow(np.flipud(img1))
    fig.add_subplot(1, 2, 2)
    pt.imshow(np.fliplr(img1))
# np.flipud() : [up/down]flips over x-axis
# np.fliplr() : [left/right]flips over y_axis
```

```python
flip(img)
```
